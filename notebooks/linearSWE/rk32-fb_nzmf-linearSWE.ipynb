# RK(3,2)-FB: Courant Number Optimization on Linearized SWEs with Non-zero Mean Flow

## Linearization of Non-linear SWEs
The nonlinear SWEs are given by
\begin{align*}
    \frac{\partial \breve{\mathbf{u}}}{\partial t} + \left( \nabla \times \breve{\mathbf{u}} + f\mathbf{k} \right) \times \breve{\mathbf{u}} &= -\nabla\frac{|\breve{\mathbf{u}}|^2}{2} - g\nabla \breve{h} \\
    \frac{\partial \breve{h}}{\partial t} + \nabla \cdot \left(\breve{h}\breve{\mathbf{u}}\right) &= 0,
\end{align*}
where $\breve{\mathbf{u}}(x, y, t) = \left(\breve{u}(x, y, t), \breve{v}(x, y, t)\right)$ is the horizontal fluid velocity, $x$ and $y$ are the spatial coordinates, $t$ is the time coordinate, $f$ is the Coriolis parameter, $g$ is the gravitational constant, $\breve{h}$ is the fluid thickness.

We can linearize these equations about a non-zero mean flow with the following assumptions:
\begin{align*}
    \breve{\mathbf{u}}(x, y, t) &= \tilde{\mathbf{U}} + \tilde{\mathbf{u}}(x, y, t) \\
    \breve{h}(x, y, t) &= H + \tilde{h}(x, y, t),
\end{align*}
where $\tilde{\mathbf{U}} = (\tilde{U}, \tilde{V})$ is the constant mean horizontal fluid velocity, $\tilde{\mathbf{u}} = \left(\tilde{u}(x, y, t), \tilde{v}(x, y, t)\right)$ is a "small" perturbation of $\tilde{\mathbf{U}}$, $H$ is the constant fluid thickness when at rest, and $\tilde{h}(x, y, t)$ is a "small" perturbation of $H$. In both cases, "small" means that the quantities themselves, as well as their derivatives, are negligible when multiplied together or squared.

Now, we make the substitutions $\breve{\mathbf{u}} = \tilde{\mathbf{U}} + \tilde{\mathbf{u}}$ and $\breve{h} = H + \tilde{h}$ and simplify term-by-term.

Kinetic energy:
\begin{align*}
    -\nabla\frac{|\breve{\mathbf{u}}|^2}{2} &= -\nabla\frac{1}{2} \left( \breve{u}^2 + \breve{v}^2 \right) \\
    &= -\nabla\frac{1}{2} \left( \tilde{U}^2 + 2\tilde{U}\tilde{u} + \tilde{u}^2 + \tilde{V}^2 + 2\tilde{V}\tilde{v} + \tilde{v}^2 \right) \\
    &= - \tilde{U}\nabla\tilde{u} - \tilde{V}\nabla\tilde{v} - \frac{1}{2}\left( \nabla\tilde{u}^2 + \nabla\tilde{v}^2 \right) \\
    &\approx - \tilde{U}\nabla\tilde{u} - \tilde{V}\nabla\tilde{v} \\
    &= \tilde{\mathbf{U}} \cdot \nabla \tilde{\mathbf{u}},
\end{align*}
where $\nabla\tilde{\mathbf{u}}$ is the Jacobian matrix $\begin{pmatrix} \partial_x\tilde{u} & \partial_x\tilde{v} \\ \partial_y\tilde{u} & \partial_y\tilde{v} \end{pmatrix}$ and $\tilde{\mathbf{U}} \cdot \nabla \tilde{\mathbf{u}}$ denotes a $2\times 1$ column vector wherein the $i$-th entry is the dot product of $\tilde{\mathbf{U}}$ with the $i$-th row of $\nabla\tilde{\mathbf{u}}$.

Potential Vorticity + Coriolis:
\begin{align*}
    \left( \nabla \times \breve{\mathbf{u}} + f\mathbf{k} \right) \times \breve{\mathbf{u}} &= \left( \left(\frac{\partial \breve{v}}{\partial x} - \frac{\partial \breve{u}}{\partial y}\right)\mathbf{k} + f\mathbf{k} \right) \times \breve{\mathbf{u}} \\
    &= \mathbf{i}\left( -\breve{v}\left( \frac{\partial \breve{v}}{\partial x} - \frac{\partial \breve{u}}{\partial y} + f \right) \right) + \mathbf{j}\left( \breve{u}\left( \frac{\partial \breve{v}}{\partial x} - \frac{\partial \breve{u}}{\partial y} + f \right) \right) \\
    &= \mathbf{i}\left( -f(\tilde{V} + \tilde{v}) - \tilde{V}\left(\frac{\partial \tilde{v}}{\partial x} - \frac{\partial \tilde{u}}{\partial y}\right) - \tilde{v}\left(\frac{\partial \tilde{v}}{\partial x} - \frac{\partial \tilde{u}}{\partial y}\right) \right) + \mathbf{j}\left( f(\tilde{U} + \tilde{u}) + \tilde{U}\left(\frac{\partial \tilde{v}}{\partial x} - \frac{\partial \tilde{u}}{\partial y}\right) + \tilde{u}\left(\frac{\partial \tilde{v}}{\partial x} - \frac{\partial \tilde{u}}{\partial y}\right) \right) \\
    &\approx \mathbf{i}\left( -f(\tilde{V} + \tilde{v}) - \tilde{V}\left(\frac{\partial \tilde{v}}{\partial x} - \frac{\partial \tilde{u}}{\partial y}\right) \right) + \mathbf{j}\left( f(\tilde{U} + \tilde{u}) + \tilde{U}\left(\frac{\partial \tilde{v}}{\partial x} - \frac{\partial \tilde{u}}{\partial y}\right) \right) \\
    &= f(\tilde{\mathbf{U}} + \tilde{\mathbf{u}})^\perp + \tilde{\zeta}\tilde{\mathbf{U}}^\perp,
\end{align*}
where $\tilde{\zeta} = \frac{\partial \tilde{v}}{\partial x} - \frac{\partial \tilde{u}}{\partial y}$ is the vertical component of the vorticity $\nabla \times \tilde{\mathbf{u}}$.

Thickness gradient:
\begin{align*}
    -g \nabla \breve{h} &= -g \nabla (H + \tilde{h}) \\
    &= -g \nabla \tilde{h}.
\end{align*}

Mass advection:
\begin{align*}
    \nabla \cdot \left( \breve{h}\breve{\mathbf{u}} \right) &= \nabla \cdot \left( H\tilde{\mathbf{U}} + H\tilde{\mathbf{u}} + \tilde{h}\tilde{\mathbf{U}} + \tilde{h}\tilde{\mathbf{u}} \right) \\
    &= \nabla \cdot \left( H\tilde{\mathbf{u}} + \tilde{h}\tilde{\mathbf{U}} + \tilde{h}\tilde{\mathbf{u}} \right) \\
    &\approx H(\nabla \cdot \tilde{\mathbf{u}}) + \nabla \cdot (\tilde{h}\tilde{\mathbf{U}}).
\end{align*}

So linearized SWEs with a non-zero mean flow are given by
\begin{align*}
    \frac{\partial \tilde{\mathbf{u}}}{\partial t} + f(\tilde{\mathbf{U}} + \tilde{\mathbf{u}})^\perp + \tilde{\zeta}\tilde{\mathbf{U}}^\perp &= -\tilde{\mathbf{U}} \cdot \nabla \tilde{\mathbf{u}} - g \nabla \tilde{h} \\
    \frac{\partial \tilde{h}}{\partial t} + H(\nabla \cdot \tilde{\mathbf{u}}) + \nabla \cdot (\tilde{h}\tilde{\mathbf{U}}) &= 0,
\end{align*}
where $\tilde{\zeta} = \frac{\partial \tilde{v}}{\partial x} - \frac{\partial \tilde{u}}{\partial y}$ .

If we take the mean flow to be zero, i.e. $\tilde{\mathbf{U}} = 0$, then the above reduces to the "standard" linearized SWEs:
\begin{align*}
    \frac{\partial \tilde{\mathbf{u}}}{\partial t} + f\tilde{\mathbf{u}}^\perp &= -g \nabla \tilde{h} \\
    \frac{\partial \tilde{h}}{\partial t} + H(\nabla \cdot \tilde{\mathbf{u}}) &= 0.
\end{align*}

## Dimensionless Equations
We can write the above in terms of dimensionless quantities by writing $c = \sqrt{gH}$, $\mathbf{u} = \frac{\tilde{\mathbf{u}}}{c}$, $\mathbf{U} = \frac{\tilde{\mathbf{U}}}{c}$, and $\eta = \frac{h}{H}$. Then, the dimensionless linearized SWEs with a non-zero mean flow are given by
\begin{align*}
    \frac{\partial \mathbf{u}}{\partial t} + f(\mathbf{U} + \mathbf{u})^\perp + c\zeta\mathbf{U}^\perp &= - c\mathbf{U} \cdot \nabla\mathbf{u} -c \nabla \eta \\
    \frac{\partial \eta}{\partial t} + c(\nabla \cdot \mathbf{u}) + c\left(\nabla \cdot (\eta\mathbf{U})\right) &= 0.
\end{align*}

## Von Neumann Formulation
With the goal of applying a von Neumann-type analysis, we apply a Fourier transform in space. This is equivalent to looking for solutions of the form
\begin{align*}
    u(x, y, t) &= \hat{u}(k, \ell, t)e^{ikx+i\ell y} \\
    v(x, y, t) &= \hat{v}(k, \ell, t)e^{ikx+i\ell y} \\
    \eta(x, y, t) &= \hat{\eta}(k, \ell, t)e^{ikx+i\ell y}.
\end{align*}

In doing this, spatial derivatives are now all of the form
$$ \frac{\partial}{\partial *} \left( e^{ikx+i\ell y} \right). $$
Assuming we are on a rectangular, staggered Arakawa C-grid, discretizing in space yields
\begin{align*}
    \frac{\partial}{\partial x} \left( e^{ikx+i\ell y} \right) &= \frac{i 2 \sin\left(k\frac{\Delta x}{2}\right)}{\Delta x}e^{ikx+i\ell y} \\
    \frac{\partial}{\partial y} \left( e^{ikx+i\ell y} \right) &= \frac{i 2 \sin\left(\ell\frac{\Delta y}{2}\right)}{\Delta y}e^{ikx+i\ell y}.
\end{align*}

To compute the (linear) Coriolis term, we need to be able to compute $u$ at $v$-points and vice-versa. On a rectangular C-grid, we do this with an unweighted 4-point average. For example, to get $v$ at a $u$-point (written as $v_u$), we calculate 
\begin{align*}
    v_u(x, y, t) &= \frac{1}{4}\hat{v}(k, \ell, t)\left( e^{ik\left(x+\frac{\Delta x}{2}\right) + i\ell\left(y+\frac{\Delta y}{2}\right)} + e^{ik\left(x+\frac{\Delta x}{2}\right) + i\ell\left(y-\frac{\Delta y}{2}\right)} + e^{ik\left(x-\frac{\Delta x}{2}\right) + i\ell\left(y+\frac{\Delta y}{2}\right)} + e^{ik\left(x-\frac{\Delta x}{2}\right) + i\ell\left(y-\frac{\Delta y}{2}\right)} \right) \\
    &= \cos\left(k\frac{\Delta x}{2}\right)\cos\left(\ell\frac{\Delta y}{2}\right)\hat{v}(k, \ell, t)e^{ikx+i\ell y}.
\end{align*}
An identical calculation holds for values of $u$ at $v$-points.

Let $\psi = f\cos\left(k\frac{\Delta x}{2}\right)\cos\left(\ell\frac{\Delta y}{2}\right)$, then the resulting equations are
\begin{align*}
    \frac{\partial \hat{u}}{\partial t} &= fV + \psi\hat{v} - \left(iU\frac{cK}{\Delta x} + iV\frac{cL}{\Delta y}\right)\hat{u} - i\frac{cK}{\Delta x}\hat{\eta} \\
    \frac{\partial \hat{v}}{\partial t} &= -fU - \psi\hat{u} - \left(iU\frac{cK}{\Delta x} + iV\frac{cL}{\Delta y}\right)\hat{v} - i\frac{cL}{\Delta y}\hat{\eta} \\
    \frac{\partial \hat{\eta}}{\partial t} &= -i\left( \frac{cK}{\Delta x}\hat{u} + \frac{cL}{\Delta y}\hat{v} \right) - i \left( U\frac{cK}{\Delta x} + V\frac{cL}{\Delta y}\right)\hat{\eta}.
\end{align*}

## Application of RK(3,2)-FB
To ease notation, we will write $u = \hat{u}$, $v = \hat{v}$, and $\eta = \hat{\eta}$ from now on. Also, write $\phi = \Delta t \psi = \Delta t f \cos\left(k\frac{\Delta x}{2}\right)\cos\left(\ell\frac{\Delta y}{2}\right)$. RK(3,2)-FB applied to this problem is then
\begin{align*}
    \bar{\eta}^{n+1/3} &= \eta^n - \frac{1}{3}\left( \left( iK\nu_x u^n + iL\nu_y v^n \right) + \left( iUK\nu_x + iVL\nu_y \right)\eta^n \right) \\
    \bar{u}^{n+1/3} &= u^n + \frac{1}{3}\left( \Delta t fV + \phi v^n - \left( iUK\nu_x + iVL\nu_y \right)u^n - iK\nu_x \eta^* \right) \\
    \bar{v}^{n+1/3} &= v^n + \frac{1}{3}\left( -\Delta t fV - \phi u^n - \left( iUK\nu_x + iVL\nu_y \right)v^n - iK\nu_x \eta^* \right) \\
    \eta^* &= \beta_1\bar{\eta}^{n+1/3} + (1-\beta_1)\eta^n \\
    & \\
    \bar{\eta}^{n+1/2} &= \eta^n - \frac{1}{2}\left( \left( iK\nu_x \bar{u}^{n+1/3} + iL\nu_y \bar{v}^{n+1/3} \right) + \left( iUK\nu_x + iVL\nu_y \right)\bar{\eta}^{n+1/3} \right) \\
    \bar{u}^{n+1/2} &= u^n + \frac{1}{2}\left( \Delta t fV + \phi \bar{v}^{n+1/3} - \left( iUK\nu_x + iVL\nu_y \right)\bar{u}^{n+1/3} - iK\nu_x \eta^{**} \right) \\
    \bar{v}^{n+1/2} &= v^n + \frac{1}{2}\left( -\Delta t fV - \phi \bar{u}^{n+1/3} - \left( iUK\nu_x + iVL\nu_y \right)\bar{v}^{n+1/3} - iK\nu_x \eta^{**} \right) \\
    \eta^{**} &= \beta_2\bar{\eta}^{n+1/2} + (1-\beta_2)\eta^n \\
    & \\
    \eta^{n+1} &= \eta^n - \left( \left( iK\nu_x \bar{u}^{n+1/2} + iL\nu_y \bar{v}^{n+1/2} \right) + \left( iUK\nu_x + iVL\nu_y \right)\bar{\eta}^{n+1/2} \right) \\
    u^{n+1} &= u^n + \left( \Delta t fV + \phi \bar{v}^{n+1/2} - \left( iUK\nu_x + iVL\nu_y \right)\bar{u}^{n+1/2} - iK\nu_x \eta^{***} \right) \\
    v^{n+1} &= v^n + \left( -\Delta t fV - \phi \bar{u}^{n+1/2} - \left( iUK\nu_x + iVL\nu_y \right)\bar{v}^{n+1/2} - iK\nu_x \eta^{***} \right) \\
    \eta^{***} &= \beta_3 \eta^{n+1} + (1-2\beta_3)\bar{\eta}^{n+1/2} + \beta_3\eta^n.
\end{align*}

```python
from sympy import *
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# define relevent symbols 
#     _u = u^n, _v = v^n, _eta = \eta^n
#     U1 = \bar{u}^{n+1/3}, V1 = \bar{v}^{n+1/3}, ETA1= \bar{\eta}^{n+1/3}
#     U2 = \bar{u}^{n+1/2}, V2 = \bar{v}^{n+1/2}, ETA2= \bar{\eta}^{n+1/2}
#     u = u^{n+1}, v = v^{n+1}, eta = \eta^{n+1}
_u, _v, _eta = symbols('u^n v^n eta^n')
lam, beta1, beta2, beta3 = symbols('lambda beta_1 beta_2 beta_3')
nux, nuy, K, L, phi, fdt = symbols('nu_x nu_y K L phi fdt')
U, V = symbols('U V')



# RK(3,2)-FB 
ETA1 = _eta - (1/3) * \
       ( (I*K*nux*_u + I*L*nuy*_v) + (I*U*K*nux + I*V*L*nuy)*_eta )
U1 = _u + (1/3) * \
     ( fdt*V + phi*_v - (I*U*K*nux + I*V*L*nuy)*_u - I*K*nux*( beta1*ETA1 + (1-beta1)*_eta ) )
V1 = _v + (1/3) * \
     ( -fdt*U - phi*_u - (I*U*K*nux + I*V*L*nuy)*_v - I*L*nuy*( beta1*ETA1 + (1-beta1)*_eta ) )

ETA2 = _eta - (1/2) * \
       ( (I*K*nux*U1 + I*L*nuy*V1) + (I*U*K*nux + I*V*L*nuy)*ETA1 )
U2 = _u + (1/2) * \
     ( fdt*V + phi*V1 - (I*U*K*nux + I*V*L*nuy)*U1 - I*K*nux*( beta2*ETA2 + (1-beta2)*_eta ) )
V2 = _v + (1/2) * \
     ( -fdt*U - phi*U1 - (I*U*K*nux + I*V*L*nuy)*V1 - I*L*nuy*( beta2*ETA2 + (1-beta2)*_eta ) )

eta = _eta - \
      ( (I*K*nux*U2 + I*L*nuy*V2) + (I*U*K*nux + I*V*L*nuy)*ETA2 )
u = _u + ( fdt*V + phi*V2 \
          - (I*U*K*nux + I*V*L*nuy)*U2 - I*K*nux*( beta3*eta + (1-2*beta3)*ETA2 + beta3*_eta ) )
v = _v + ( -fdt*U - phi*U2 \
          - (I*U*K*nux + I*V*L*nuy)*V2 - I*L*nuy*( beta3*eta + (1-2*beta3)*ETA2 + beta3*_eta ) )
```

## Amplification matrices

Now, we have a system of linear equations in $\eta^n$, $u^n$, and $v^n$. We can find the amplification matrix(??) for the scheme by writing it in the form
$$ \mathbf{w}^{n+1} = G \mathbf{w}^n + b, $$
where $G$ is the amplification matrix, $\mathbf{w}^n = (u^n, v^n, \eta^n)^T$ and $b$ is some constant column vector.

This allows us to write
\begin{align*}
    \mathbf{w}^n &= G^n \mathbf{w}^0 + \left(\sum_{j=0}^{n-1}G^j\right)b \\
    &= G^n \mathbf{w}^0 + (I- G^n)(I - G)^{-1} b.
\end{align*}
It follows that $||\mathbf{w}^n||$ is bounded if and only if $||G^n||$ is bounded.

Normally, we would now be interested in the eigenvalues of $G$ given that it can be shown that as long as $G$ is diagonalizable, then $||G^n||$ is bounded if and only if each eigenvalue $\lambda$ of $G$ is such that $|\lambda| \leq 1$. However, we are first interested in choosing the "optimal" FB feedback coefficients $\beta_i$.

To do this, we first need to know the eigenvalues of the amplification matrix in the continuous problem. The continuous problem can be written as
$$ \frac{\partial}{\partial t} \begin{pmatrix}
    u \\ v \\ \eta
\end{pmatrix} = \begin{pmatrix}
    -\left(iUck + iVc\ell\right) & f & -ick \\ 
    -f & -\left(iUck + iVc\ell\right) & -ic\ell \\ 
    -ick & -ic\ell & -\left(iUck + iVc\ell\right)
\end{pmatrix} \begin{pmatrix}
    u \\ v \\ \eta
\end{pmatrix} + \begin{pmatrix}
    fV \\ -fU \\ 0
\end{pmatrix}. $$

It can be shown that the general solution to the system is
$$ \begin{pmatrix}
    u(t) \\ v(t) \\ \eta(t)
\end{pmatrix} = e^{ At }\begin{pmatrix}
    u(0) \\ v(0) \\ \eta(0)
\end{pmatrix} + \int_0^t e^{(t - \tau)A}b\ dt, $$
where
$$ A = \begin{pmatrix}
    -\left(iUck + iVc\ell\right) & f & -ick \\ 
    -f & -\left(iUck + iVc\ell\right) & -ic\ell \\ 
    -ick & -ic\ell & -\left(iUck + iVc\ell\right)
\end{pmatrix} $$
and
$$ b = \begin{pmatrix}
    fV \\ -fU \\ 0
\end{pmatrix}. $$

Now, consider the general solution of the system after exactly one time-step, i.e. at  time $t = \Delta t$. Then,
\begin{align*}
    A \Delta t &= \begin{pmatrix}
        -\Delta t\left(iUck + iVc\ell\right) & f \Delta t & -ic\Delta t k \\ 
        -f \Delta t & -\Delta t\left(iUck + iVc\ell\right) & -ic\Delta t \ell \\ 
        -ic\Delta t k & -ic\Delta t\ell & -\Delta t\left(iUck + iVc\ell\right)
    \end{pmatrix} \\
    &= \begin{pmatrix}
        -\left(iU\tilde{K}\nu_x + iV\tilde{L}\nu_y\right) & f \Delta t & -i\tilde{K}\nu_x \\ 
        -f \Delta t & -\left(iU\tilde{K}\nu_x + iV\tilde{L}\nu_y\right) & -i\tilde{L}\nu_y \\ 
        -i\tilde{K}\nu_x & -i\tilde{L}\nu_y & -\left(iU\tilde{K}\nu_x + iV\tilde{L}\nu_y\right)
    \end{pmatrix}.
\end{align*}
where $\tilde{K} = k \Delta x$ and $\tilde{L} = \ell \Delta y$. The final step in the above is done so that we can have the Courant numbers $\nu_x$ and $\nu_y$ baked directly into the notation.

Let $\tilde{G} = e^{A\Delta t}$. This is the exact amplification matrix we wish to "approximate" by our numerical amplification matrix $G$.

```python
# We build Gtilde using scipy, not sympy -- the symbolic calculation
# of e^{A\Delta t} takes too long (~4 hours). The symbolic expression 
# is too complicated to be meaningful to a human anyway.
def _calcGtilde(nux, nuy, Ktilde, Ltilde, fdt, U, V):
    Adt = np.matrix( [[-(1j*U*Ktilde*nux + -1j*V*Ltilde*nuy), fdt, -1j*Ktilde*nux],
                      [-fdt, -(1j*U*Ktilde*nux + -1j*V*Ltilde*nuy), -1j*Ltilde*nuy],
                      [-1j*Ktilde*nux, -1j*Ltilde*nuy, -(1j*U*Ktilde*nux + -1j*V*Ltilde*nuy)]] )
    return sp.linalg.expm(Adt)
# END calcGtilde()
```

## Statement of the optimization problem

Now we can clearly state our goal: Choose the FBFCs $\beta_i$ so that the eigenvalues of the numerical amplification matrix $G$ approximate the eigenvalues of $\tilde{G} = e^{A\Delta t}$ while maximizing the Courant numbers $\nu_x$ and $\nu_y$. To do this, we are going to solve a numerical optimization problem.

Let $\nu_x^\text{max}$ and $\nu_y^\text{max}$ be the maximum allowable Courant numbers for stability of the method. These depend on our choice of the FBFCs $\beta_i$. We can maximize
$\nu_*^\text{max}$ by minimizing an appropriate cost function
$$ C = C(\nu_x^\text{max}, \nu_y^\text{max}, \beta_1, \beta_2, \beta_3), $$
subject to the restriction that eigenvalues of $G$, $\lambda = \lambda(\nu_x, \nu_y, \beta_1, \beta_2, \beta_3)$, are such that
$$ |\lambda(\nu_x, \nu_y, \beta_1, \beta_2, \beta_3)| \leq 1 \qquad \forall\ (\nu_x, \nu_y) \in (0, \nu_x^\text{max}] \times (0, \nu_y^\text{max}]. $$

### Simplifying assumptions

Before we can move forward, we need to address the additional parameters that both $\tilde{G}$ and $G$ depend on. These are:
\begin{align*}
    \tilde{K} &= k\Delta x \\
    \tilde{L} &= \ell\Delta y \\
    & \\
    K &= 2\sin\left(k\frac{\Delta x}{2}\right) \\
    L &= 2\sin\left(\ell\frac{\Delta y}{2}\right) \\
    & \\
    \phi &= f\Delta t \cos\left(k\frac{\Delta x}{2}\right)\cos\left(\ell\frac{\Delta y}{2}\right)
\end{align*}
We need to fix these values so that they are not counted as free parameters in the optimization problem. As can be seen in the cell below, we only need to fix values for $\tilde{K}$ and $f\Delta t$, then values for the other parameters follow.

According to Bob, a decent choice for $f$ would be $f = 10^{-4}$, and since the time-steps for the nonlinear SWE test cases we are looking at are on the order of $10^2$, a good choice for $f\Delta t$ is
$$f\Delta t = 10^{-2}.$$

For $\tilde{K}$ and $\tilde{L}$, we want to sample this space:
$$\left(\tilde{K}, \tilde{L} \right) \in [0, \pi] \times [0, \pi].$$
For each pair $\left(\tilde{K}, \tilde{L} \right)$, we need to complete the optimization problem and get a suggested tuple of $\beta_i$.

```python
# Define a function to sort eigenvalues so we know which to compare
def parseEigVals(eigVals):
    eigValDict = {}
    reals = []
    for eigVal in eigVals:
        if np.isclose(eigVal, 1.0+0.0j):
            eigValDict['one'] = eigVal
        elif not np.isclose(eigVal.imag, 0.0):
            if eigVal.imag > 0:
                eigValDict['pos'] = eigVal
            else:
                eigValDict['neg'] = eigVal
        else:
            reals.append(eigVal.real)
            eigValDict['real'] = reals
        # END if
    # END for
    return eigValDict
#END parseEigVals()


# Define a function to create a plot comparing exact/numerical eigenvalues
def eigValPlot(Gtilde, G, KtildeVal=-1, LtildeVal=-1,
               a=-16, b=16, delta=1e-2, nuDotLimit=np.pi/3):
    
    red = '#ee4035'
    orange = '#f37736'
    yellow = '#fdf498'
    green = '#7bc043'
    blue = '#0392cf'
    
    fig, ax = plt.subplots(1)

    ax.spines['left'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('center')
    ax.spines['top'].set_color('none')

    axLim = 1.5
    ax.set_xlim(-axLim, axLim)
    ax.set_ylim(-axLim, axLim)
    ax.set_aspect(1)

    unitCircle = plt.Circle((0, 0), 1, fill=False, color=orange)
    ax.add_artist(unitCircle)

    approx_real = []
    approx_imag = []
    badNus =[]
    count = 0

    for k in np.arange(a, b, delta):
        nu = k*np.pi/16
        # get complex values for exact and approx eigen values
        GtildeEigVals = parseEigVals( sp.linalg.eigvals( Gtilde(nu) ) )
        GEigVals = parseEigVals( sp.linalg.eigvals( G(nu) ) )

        # this process needs to be revisited
        if np.isclose(nu, 0):
            exact = 1+0.j
            approx = 1.+0.j
        else:
            key = 'pos' if nu > 0 else 'neg'
            exact = GtildeEigVals[key]
            if 'real' not in GEigVals.keys():
                approx = GEigVals[key]
            else:
                if len(GEigVals['real']) == 1:
                    approx = GEigVals['real'][0]
                else:
                    ind = 1 if nu >= 0 else 0
                    approx = GEigVals['real'][ind]
                # END if
            # END if
        # END if
        approx_real.append( approx.real )
        approx_imag.append( approx.imag )
        
        absValApprox = sqrt(approx.real**2 + approx.imag**2)
        if absValApprox > 1: badNus.append(nu) 
        
        if ( np.isclose(k%1, 0) or np.isclose(k%1, 1) ) and np.abs(nu) <= nuDotLimit:
            # plot exact and approx eigen values
            colo = blue if absValApprox <= 1 else red
            ax.scatter(exact.real, exact.imag, marker='o', color=orange)
            ax.scatter(approx.real, approx.imag, marker='*', color=colo)
            # connect exact to approx with dashed line
            reals = [approx.real, exact.real]
            imags = [approx.imag, exact.imag]
            ax.plot(reals, imags, color='black', linestyle='dashed', linewidth=0.75, alpha=0.5)
        # END if
    # END for

    ax.plot(approx_real, approx_imag, color=blue)
    try:
        print( 'nu^max ~ {}'.format( np.min(np.abs(badNus)) ) )
    except: pass

# END eigValPlot()

def eigValPlotSimple(Gtilde, G, KtildeVal=-1, LtildeVal=-1,
                     a=-16, b=16, delta=1e-2, nuDotLimit=np.pi/3):
    
    red = '#ee4035'
    orange = '#f37736'
    yellow = '#fdf498'
    green = '#7bc043'
    blue = '#0392cf'
    
    fig, ax = plt.subplots(1)

    ax.spines['left'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('center')
    ax.spines['top'].set_color('none')

    axLim = 1.5
    ax.set_xlim(-axLim, axLim)
    ax.set_ylim(-axLim, axLim)
    ax.set_aspect(1)

    unitCircle = plt.Circle((0, 0), 1, fill=False, color=orange)
    ax.add_artist(unitCircle)

    approx_real = []
    approx_imag = []
    badNus =[]
    count = 0

    for k in np.arange(a, b, delta):
        nu = k*np.pi/16
        # get complex values for exact and approx eigen values
        GtildeEigVals = sp.linalg.eigvals( Gtilde(nu) )
        GEigVals = sp.linalg.eigvals( G(nu) )
        
        for eigVal in GEigVals:
            absValApprox = sqrt(eigVal.real**2 + eigVal.imag**2)
            if absValApprox > 1: badNus.append(nu)

            ax.scatter(eigVal.real, eigVal.imag, marker='*', color=blue)
        # END for
        for eigVal in GtildeEigVals:
            ax.scatter(eigVal.real, eigVal.imag, marker='o', color=orange)
        # END for
    # END for
    try:
        print( 'nu^max ~ {}'.format( np.min(np.abs(badNus)) ) )
    except: pass
# END eigValPlotSimple()
```

## Setup for the optimization problem

Recall that the primary constraint on our optimization problem is that
$$ |\lambda(\nu_x, \nu_y, \beta_1, \beta_2, \beta_3)| \leq 1 \qquad \forall\ (\nu_x, \nu_y) \in (0, \nu_x^\text{max}] \times (0, \nu_y^\text{max}]. $$

As described above, we are using the simplifying assumption that $\nu_x = \nu_y = \nu$, so this becomes
$$ |\lambda(\nu, \beta_1, \beta_2, \beta_3)| \leq 1 \qquad \forall\ \nu \in (0, \nu^\text{max}]. $$
In practice, we check this condition by checking the absolute value of the eigenvalues on a grid in $\nu$, i.e. we check
$$ |\lambda(\nu, \beta_1, \beta_2, \beta_3)| \leq 1 \qquad \forall \nu = \delta, 2\delta, 3\delta, \cdots \nu^\text{max}, $$
where $\delta$ is a user-defined step-size in `eigValMax()` (defined below), the smaller $\delta$ is, the longer the numerical optimization takes to execute.

```python
# Require that |lambda| <= 1 for all 0 < nu <= nu^max
def eigValMax(x, delta=np.pi/512): # x = [nu_max, beta1, beta2, beta3]
    global G

    absList = []
    
    nus = np.arange(delta, x[0], delta)
    if x[0] not in nus: nus = np.append(nus, x[0])
    
    for nu in nus:
        eigVals = sp.linalg.eigvals( G(nu, x[1], x[2], x[3]) )
        for eigVal in eigVals:
            absVal = np.abs(eigVal)
            if not np.isclose(absVal, 1.0):
                absList.append( absVal )
            else:
                absList.append(1.0)
        # END for
    # END for
    return np.max(absList)
# END eigValMax()

# Set up the contraints for scipy.optimize.shgo()
eigValCons = lambda x: 1 - eigValMax(x) # >= 0
cons = [{'type': 'ineq', 'fun': eigValCons}]
```

### A cost function

Our first choice of cost function will be
$$ C_1(\nu^\text{max}, \beta_1, \beta_2, \beta_3) = \frac{1}{\nu^\text{max}} + \int_{\frac{-\pi}{6}}^{\frac{\pi}{6}} \left| \lambda_\text{exact}(\nu) - \lambda(\nu, \beta_1, \beta_2, \beta_3) \right| \text{d}\nu. $$
The first term asks that $\nu^\text{max}$ be as large as possible, while the second term asks that, over the given bounds, the absolute difference between the eigenvalues of $\tilde{G}$ and $G$ be as small as possible.

In practice, the integral term in the above will be evaluated as a simple sum (though we could consider doing a quadrature of some sort if we really wanted to). This is implemented below in `eigValDiffSum()`.

```python
# Define a function that gives a measure of how close the numerical eigenvalues are
# to their exact counterparts, we want this to be small for given nu_max and beta_i
def eigValDiffSum(x, G, Gtilde, b=np.pi/3, delta=np.pi/64):   
    diffSum = 0
    
    nus = np.arange(0, b, delta)
    for nu in nus:
        GtildeEigVals = [eigVal for eigVal in \
                         sp.linalg.eigvals( Gtilde(nu) ) \
                         if not np.isclose(eigVal, 1.+0.j)]
        GEigVals = [eigVal for eigVal in \
                    sp.linalg.eigvals( G(nu, x[1], x[2], x[3]) ) \
                    if not np.isclose(eigVal, 1.+0.j)]
        
        if np.isclose(nu, 0):
            diffSum += 0
        else:
            diffs = []
            for GEigVal in GEigVals:
                for GtildeEigVal in GtildeEigVals:
                    diffs.append( np.abs(GEigVal - GtildeEigVal) )
                # END for
            # END for
            diffSum += np.min(diffs)
        # END if
    # END for
    return 2*delta*diffSum
# END eigValDiffSum()
```

Now, we can use `scipy.optimize.shgo()` to find a numerical solution to the optimization problem stated above. This is a global optimization scheme and seems to be much more robust than `scipy.optimize.minimize()` which only acts as a local optimizer.

```python
res = {}
```

```python
fdtVal = 1e-2
UVal = 80 / np.sqrt(9.8*10000)
VVal = UVal

KtildeVal = np.pi
LtildeVal = np.pi

KVal = 2*np.sin(KtildeVal/2)
LVal = 2*np.sin(LtildeVal/2)
phiVal = fdtVal*np.cos(KtildeVal/2)*np.cos(LtildeVal/2)

G = lambda nu, beta1, beta2, beta3: _calcG(nu, nu, beta1, beta2, beta3, 
                                           KVal, LVal, phiVal, UVal, VVal)
Gtilde = lambda nu: _calcGtilde(nu, nu, KtildeVal, LtildeVal, fdtVal, UVal, VVal)

cost = lambda x: 1/x[0] + eigValDiffSum(x, G, Gtilde, b=0)
bounds = [(1e-2, 2), (0, 1), (0, 1), (0, 1)]
res['C1'] = sp.optimize.shgo(cost, bounds, constraints=cons, iters=6)
res['C1']
```

```python
vals = res['C1'].x
optG = lambda nu: G(nu, vals[1], vals[2], vals[3])

eigValPlotSimple(Gtilde, optG, nuDotLimit=0)
```
